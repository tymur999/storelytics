```python
from deepface import DeepFace
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("./data/tymur.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
obj = DeepFace.analyze(image, actions = ['race', 'emotion', 'gender'])
```
